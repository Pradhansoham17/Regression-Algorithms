```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_boston
X, y = load_boston(return_X_y=True)
load_boston()
```

```python
X
```

```python
y
```

```python
load_boston()['feature_names']
```

```python
df=pd.DataFrame(data=X, columns=load_boston()['feature_names'])
df['house-price']=y
df
```

```python
from sklearn.model_selection import train_test_split
X, y = df.iloc[:,:-1],df.iloc[:,[-1]]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=5)
```

```python
from sklearn.preprocessing import StandardScaler
data = X_train
scaler = StandardScaler()
```

```python
X_train = pd.DataFrame(scaler.fit_transform(data),index=X_train.index, columns=X_train.columns)
X_train
```

```python
X_test = pd.DataFrame(scaler.transform(X_test),index=X_test.index, columns=X_test.columns)
X_test
```

```python
#without train test split
from sklearn.linear_model import Ridge
clf = Ridge(alpha=1.0, max_iter=1000)
clf.fit(X_train, y_train)
```

```python
clf.coef_
```

```python
clf.intercept_
```

```python
clf.predict(X_test)
```

```python
def MeanSquaredError(Y_true,Y_pred):
    MSE = np.square(np.subtract(Y_true,Y_pred)).mean() 
    return MSE

Y_true = [1,1,2,2,4]
Y_pred = [0.6,1.29,1.99,2.69,3.4]

MeanSquaredError(Y_true,Y_pred)   
```

```python
MeanSquaredError(y_test,clf.predict(X_test))
```
