```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_boston
X, y = load_boston(return_X_y=True)
load_boston()
```

```python
X
```

```python
y
```

```python
load_boston()['feature_names']
```

```python
df=pd.DataFrame(data=X, columns=load_boston()['feature_names'])
df['house-price']=y
df
```

```python
from sklearn.model_selection import train_test_split
X, y = df.iloc[:,:-1],df.iloc[:,[-1]]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=5)
```

```python
X_train
```

```python
X_test
```

```python
from sklearn.preprocessing import StandardScaler
data = X_train
scaler = StandardScaler()
```

```python
scaler.fit_transform(data)
```

```python
X_train = pd.DataFrame(scaler.fit_transform(data),index=X_train.index, columns=X_train.columns)
X_train
```

```python
X_test = pd.DataFrame(scaler.transform(X_test),index=X_test.index, columns=X_test.columns)
X_test
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression().fit(X_train, y_train)
reg.score(X_train, y_train) #returns the r-square coefficient of train data
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
y_pred=reg.predict(X_test)
y_pred
```

```python
# for comparing predicted and actual values side by side
np.concatenate((y_pred.reshape(len(y_pred),1), y_test.values.reshape(len(y_test),1)),1)
```

```python
reg.score(X_test, y_test)
```
