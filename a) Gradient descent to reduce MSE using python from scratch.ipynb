# Updating the values of 'm' and 'c' till fixed number of iterations :-

```python
import pandas as pd
import numpy as np
df=pd.read_csv('C:\\Users\\soham\\Downloads\\LinearRegression.csv')
df 
```

```python
def MeanSquaredError(Y_true,Y_pred):
    MSE = np.square(np.subtract(Y_true,Y_pred)).mean() 
    return MSE

Y_true = [1,1,2,2,4]
Y_pred = [0.6,1.29,1.99,2.69,3.4]

MeanSquaredError(Y_true,Y_pred)   
```

```python
def gradient_descent(X, y_true, lr=0.05, epoch=100): #updating m and c till fixed number of iteration/epoch,by default epoch=100
    
    '''
    Gradient Descent for a single feature
    '''
    
    m, c = 3,2 # parameters
    logs, mse = [], [] # lists to store learning process
    N = len(X) # number of samples
    
    for i in range(epoch):
        
        y_pred = (m*X + c)
        diff = y_pred - y_true 
        
        MSE= MeanSquaredError(y_true, y_pred)
        print('The Mean Squared Error of epoch ',i,' is ',MSE)
        
        logs.append((m, c))
        mse.append(MSE)
        
        # lists for plotting purposes
        mse_list.append(MSE)
        m_list.append(m)
        c_list.append(c)
        
        m -= lr * (2 * np.dot(X,diff).sum() / N)  # acc to gradient descent eqn, this is m(new) = (m(old) + α * ∂(MSE)/∂m ) , where α is learning rate
        c -= lr * (2 * diff.sum() / N)  # acc to gradient descent eqn, this is c(new) = (c(old) + α * ∂(MSE)/∂c ) , where α is learning rate
        
    print('The Final Value of m is : ',m)
    print('The Final Value of c is : ',c)
    print('The Minimum Value of MSE is : ',min(mse))
    
           
    return logs, mse
```

```python
# lists for plotting purposes
mse_list=[]
m_list=[]
c_list=[]

X=df['x']
y=df['y']
gradient_descent(X, y, lr=0.05, epoch=1000)
```

```python
# importing the required module 
import matplotlib.pyplot as plt 

# x axis values 
x = m_list

# corresponding y axis values 
y = mse_list

# plotting the points 
plt.plot(x, y) 

# naming the x axis 
plt.xlabel('m') 
# naming the y axis 
plt.ylabel('Mean Squared Error') 

# giving a title to my graph 
plt.title('Mean Squared Error vs m Graph !!') 

# function to show the plot 
plt.show() 
```

# Updating the values of 'm' and 'c' till MSE(next) > MSE(previous) :-

```python
def gradient_descent(X, y_true, lr=0.05, epoch=100):
    
    '''
    Gradient Descent for a single feature
    '''
    
    m, c = 3,2 # parameters
    logs, mse = [], [] # lists to store learning process
    N = len(X) # number of samples
    iteration = 0
    
    while(len(mse)<2 or mse[-1] < mse[-2]): 
        
        iteration += 1
            
        # initially 'mse' is an empty list hence mse[-1] or mse[-2] doesn't exist thus it may give error, 
        # that's why the condition len(mse)<2, so that comparison gets started only after first 2 elements 
        # of the list are filled.
        
        y_pred = (m*X + c)
        diff = y_pred - y_true 
        
        MSE= MeanSquaredError(y_true, y_pred)
        print('The Mean Squared Error is currently ',MSE)
        
        logs.append((m, c))
        mse.append(MSE)
        
        m -= lr * (2 * np.dot(X,diff).sum() / N)  # acc to gradient descent eqn, this is m(new) = (m(old) + α * ∂(MSE)/∂m ) , where α is learning rate
        c -= lr * (2 * diff.sum() / N)  # acc to gradient descent eqn, this is c(new) = (c(old) + α * ∂(MSE)/∂c ) , where α is learning rate
        
        # lists for plotting purposes
        mse_list.append(MSE)
        m_list.append(m)
        c_list.append(c)
        
    print('The Final Value of m is : ',m)
    print('The Final Value of c is : ',c)
    print('The Minimum Value of MSE is : ',min(mse))
    print('The number of iterations occured are : ',iteration)
    
           
    return logs, mse
```

```python
# lists for plotting purposes
mse_list=[]
m_list=[]
c_list=[]

X=df['x']
y=df['y']
gradient_descent(X, y, lr=0.05, epoch=1000)
```

```python
# plot will update itself in the same axes as we go ahead
# https://stackoverflow.com/questions/34486642/what-is-the-currently-correct-way-to-dynamically-update-plots-in-jupyter-ipython

import numpy as np
import matplotlib.pyplot as plt
import time

def pltsin(ax, colors=['b']):
    x = np.linspace(0,1,100)
    if ax.lines:
        for line in ax.lines:
            line.set_xdata(x)
            y = np.random.random(size=(100,1))
            line.set_ydata(y)
    else:
        for color in colors:
            y = np.random.random(size=(100,1))
            ax.plot(x, y, color)
    fig.canvas.draw()

fig,ax = plt.subplots(1,1)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_xlim(0,1)
ax.set_ylim(0,1)
for f in range(5):
    pltsin(ax, ['b', 'r'])
    time.sleep(1)
```
